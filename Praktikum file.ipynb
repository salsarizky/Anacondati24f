```python
# data_mahasiswa = []

def tambah_mahasiswa():
    print("\n--- Tambah Data Mahasiswa ---")
    while True:
        nim = input("Masukkan NIM (11 digit): ")
        if len(nim) == 11 and nim.isdigit():
            # Cek apakah NIM sudah ada
            if any(m['nim'] == nim for m in data_mahasiswa):
                print("NIM sudah terdaftar. Mohon masukkan NIM yang berbeda.")
            else:
                break
        else:
            print("NIM harus 11 digit angka. Silakan coba lagi.")

    while True:
        nama = input("Masukkan Nama (minimal 6 karakter): ")
        if len(nama) >= 6:
            break
        else:
            print("Nama minimal 6 karakter. Silakan coba lagi.")

    while True:
        alamat = input("Masukkan Alamat (minimal 10 karakter): ")
        if len(alamat) >= 10:
            break
        else:
            print("Alamat minimal 10 karakter. Silakan coba lagi.")

    while True:
        kelas = input("Masukkan Kelas (TI23/TI24): ").upper()
        if kelas in ["TI23", "TI24"]:
            break
        else:
            print("Kelas hanya boleh 'TI23' atau 'TI24'. Silakan coba lagi.")

    mahasiswa = {
        "nim": nim,
        "nama": nama,
        "alamat": alamat,
        "kelas": kelas
    }
    data_mahasiswa.append(mahasiswa)
    print("Data mahasiswa berhasil ditambahkan!")

def tampilkan_mahasiswa():
    print("\n--- Daftar Data Mahasiswa ---")
    if not data_mahasiswa:
        print("Belum ada data mahasiswa.")
    else:
        for i, m in enumerate(data_mahasiswa):
            print(f"\nMahasiswa ke-{i+1}:")
            print(f"  NIM    : {m['nim']}")
            print(f"  Nama   : {m['nama']}")
            print(f"  Alamat : {m['alamat']}")
            print(f"  Kelas  : {m['kelas']}")

def main_menu():
    while True:
        print("\n--- Menu Program Data Mahasiswa ---")
        print("1. Tambah Data Mahasiswa")
        print("2. Tampilkan Data Mahasiswa")
        print("3. Keluar")

        pilihan = input("Pilih menu (1/2/3): ")

        if pilihan == '1':
            tambah_mahasiswa()
        elif pilihan == '2':
            tampilkan_mahasiswa()
        elif pilihan == '3':
            print("Terima kasih! Program selesai.")
            break
        else:
            print("Pilihan tidak valid. Silakan coba lagi.")

# Jalankan program
if __name__ == "__main__":
    main_menu()
 
```
